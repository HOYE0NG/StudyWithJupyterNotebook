### 데이터베이스에서 인덱스를 사용하는 이유 및 장단점에 대해 설명하시오.
검색 성능을 향상시키기 위해 컬럼을 기준으로 정렬해놓은 복사본을 인덱스라고 한다. 
단점
- 컬럼을 복사해서 정렬하는 것이기 때문에 DB 용량을 차지한다. 
- 삽입 수정 삭제 시 인덱스에도 반영해야하는데 이 때 오버헤드가 발생
이진 검색 트리를 성능 강화시킨 B+ Tree 를 사용한다. 

### 트랜잭션에 대해 설명하시오.
트랜잭션이란 데이터베이스의 상태를 변화시키는 하나의 논리적인 작업 단위이다. 트랜잭션 수행 중에 하나의 작업이라도 실패하면 전부 실패하고, 모두 성공해야 성공이라고 할 수 있다.

### ACID에 대해 설명하시오.
트랜잭션의 신뢰성을 보장하기 위한 네가지 속성
Atomicity(원자성) : 트랜잭션의 연산은 모든 연산이 완벽히 수행되어야 하며, 한 연산이라도 실패하면 트랜잭션의 모든 연산이 실패되어야 한다.
Consistency(일관성) : 트랜잭션이 완료된 후 데이터베이스가 일관된 상태를 유지하도록 해야 한다. 유효한 상태를 유지해야 한다. 무결성 제약 조건을 항상 만족시켜야 한다.
Isolation(고립성) : 트랜잭션은 다른 트랜잭션에 의해 영향받지 않고 독립적으로 실행되어야 한다.
Durability(내구성) : 트랜잭셔니 커밋된 이후에는 시스템 오류가 발생하더라도 커밋된 상태로 유지되는 것이 보장되어야 한다. (비휘발성 메모리에 저장되어야 한다.)

### 트랜잭션 격리 수준(Transaction Isolation Level)에 대해 설명하시오
여러 트랜잭션이 동시에 처리될 때, 특정 트랜잭션이 다른 트랜잭션에서 변경하거나 조회하는 데이터를 볼 수 있게 허용할지 여부를 결정하는 것. 각 격리 수준은 고립도와 성능사이의 트레이드 오프를 조절하며, 부정합(더티 리드, 반복되지 않는 리드, 팬텀 리드) 문제들의 발생 가능성을 다룬다.

- READ UNCOMMITTED : 커밋되지 않은 변경 내용도 참조 가능 -> 더티 리드(롤백한 것 읽어버리기)
- READ COMMITTED : 커밋된 내용만 참조 가능 -> 반복되지 않는 읽기(동일 트랜잭션 내에서 다시 읽을 때 다른 값)
- REPEATABLE READ : 트랜잭션 진입 이전에 커밋된 내용만 참조 가능 -> 팬텀 리드(처음에 없던 데이터가 나탈 수 있음) 
- SERIALIZABLE : 트랜잭션을 순차적으로 진행시켜서 레코드 동시 접근 방지

### 무결성 제약 조건에 대해 설명하시오.
도메인 무결성 : 각 컬럼에 저장되는 데이터의 타입과 형식을 정의해야 한다.
개체 무결성 : 모든 테이블은 기본 키가 존재하고 기본키가 각 행을 고유하게 식별해야한다. 기본키가 null 이면 안되고 중복되어도 안된다.
참조 무결성 : 외래키로 참조되어지고 있는 키가 존재해야 한다.

### 정규화에 대해 설명하시오.
데이터베이스의 이상현상을 해결하기 위해 테이블을 쪼개는 것. 조인 많아져서 성능 떨어지면 반정규화하기도 한다.
1차 : 컬럼의 원자성 보장
2차 : 부분 함수 종속성 제거 (기본키의 부분 집합이 다른 속성의 결정자가 되면 안된다)
3차 : 이행적 함수 종속성 제거 (A->B->C와 같은 종속성 제거)
BCNF : 기본키가 아닌 다른 속성들이 결정자되면 안됨
4차 : 다치 종속 제거
5차 : 조인 종속성 제거

### JOIN에 대해 설명하시오.
두개 이상의 테이블을 결합해서 결과 집합을 생성하기 위한 연산. 조인 연산이 많아지면 성능이 떨어진다.

### RDBMS vs NoSQL
RDBMS, 관계형 데이터베이스는 말그대로 객체들의 릴레이션, 테이블을 통해서 2차원 구조로 데이터를 저장하는 방식의 데이터베이스이다. 
장점 : 복잡한 데이터를 다루기 쉽고 트랜잭션을 통해 안정적 데이터 관리 가능
단점 : 정해진 스키마에 맞춰서 데이터를 다뤄야함. + 유연성이 낮아 성능 향상을 위해서는 스케일업(업그레이드)만 가능해서 비용이 크다.
NoSQL은 비관계형 데이터베이스로 사전에 스키마를 정의하지 않아도 되기 때문에 다양한 형식의 데이터를 쉽게 처리 가능하고 Scale-out(데이터베이스 추가)이 가능해 유연성이 높다. 하지만 데이터 변형, 중복을 막기 어렵다.

### MongoDB에 대해 설명하시오
문서지향 NoSQL 데이터베이스로 비정형 데이터를 JSON형식으로 저장한다. 데이터를 저장하는 문서와 문서들의 집합인 컬렉션으로 구성된다.

### Redis에 대해 설명하시오.
key-value 타입의 인메모리 데이터베이스로 RAM에 데이터를 저장하고 조회하기 때문에 속도가 빠르지만 휘발성으로 인해 용량 문제로 데이터 유실이 발생할 수 있다. 따라서 SUB DB로 많이 사용된다.
