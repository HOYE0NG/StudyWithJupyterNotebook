### 프로세스와 쓰레드의 차이에 대해 설명하시오.
프로세스는 실행 중인 프로그램, 운영체제로부터 자원을 할당받은 작업의 단위를 의미한다. 스레드는 프로세스가 할당받은 자원을 이용하는 실행 흐름의 단위이다. 각각의 프로세스는 독립적인 메모리 영역을 가지기 때문에 자원 공유에 대한 동기화 문제가 스레드에 비해 없지만 컨텍스트 스위칭 오버헤드가 크다. 쓰레드는 스택 영역을 제외한 메모리 영역을 프로세스 내부에서 서로 공유하기 때문에 컨텍스트 스위칭 오버헤드가 적지만 레이스 컨디션과 같은 동기화 문제가 발생할 수 있다.

### 컨텍스트 스위칭에 대해 설명하시오.
한 작업이 끝날 때까지 기다리는 것이 아니라 커널이 여러 작업을 번갈아가며 실행해서 동시에 처리 될 수 있도록 하는 방법이다. 인터럽트가 발생하면, 현재 프로세스의 상태를 PCB에 저장하고 새로운 프로세스의 상태를 레지스터에 저장하는 방식으로 동작한다. 이 과정에서 cpu는 아무런 작업을 하지 않으므로 컨텍스트 스위칭이 잦으면 성능 저하를 일으킬 수 있다. 

- 동시성(Concurrency)과 병렬성(Parallelism)
동시성은 논리적으로 동시에 실행한다는 뜻으로 여러 작업을 번갈아가며 수행하는 것이다. 병렬성은 물리적으로 동시에 실행한다는 뜻으로 실제로 여러 작업을 동시에 실행하는 것이다.
- 동기화 문제
1. Race condition - 경쟁 상태
여러 스레드가 동시에 공유 자원에 접근하고 수정하려 할 때 발생하는 것. -> 뮤텍스, 세마포어로 해결
2. DeadLock - 교착 상태
두 개 이상의 스레드가 서로의 자원을 기다리며 무한 대기 상태에 빠지는 것.(상태 안바꿈) -> 자원 할당 순서 고정 or 타임아웃 설정
3. LiveLock 
두 개 이상의 스레드가 락의 해제와 획득을 무한 반복하여 어느 쓰레드도 자원을 할당받지 못하는 것.(상태를 계속 바꿈) -> 스레드가 랜덤 시간에 자원 점유 재시도하도록 함.
4. Starvation - 기아 상태
우선순위가 높은 스레드가 자원을 점유해서 낮은 스레드가 자원을 할당받지 못하고 대기하는 것. -> 우선순위를 수시로 변경

### 동기와 비동기의 차이 (블로킹, 논블로킹) 에 대해 설명하시오.
동기는 한 작업이 끝나면 다음 작업을 진행하는 순차적인 방식(블로킹: 한작업이 완료 될 때 까지 기다림)이고 비동기는 끝날 때 까지 기다리지 않고 비순차적으로 실행하는 방식(논블로킹: 한작업이 완료 될 때까지 안기다림). 동기 방식은 코드가 순차적으로 진행되므로 이해하기 쉽고 디버깅이 용이하지만 단위 시간당 비교적 적은 작업을 처리한다. CPU가 IO 작업이 끝날 때까지 기다리므로 비효율적. 비동기 방식은 단위시간당 많은 작업을 저리할 수 있지만(IO 작업에 유리) 복잡성이 증가하고 디버깅이 어렵다.

### 멀티 프로세스, 멀티 스레드 프로그래밍에 대해 설명하시오.
멀티 프로세스는 여러 프로세스로 작업을 병렬로 처리하는 방식이고 멀티 스레드는 하나의 프로세스에서 여러 개의 스레드들을 만들어 처리하는 방식. 멀티 프로세스는 각각의 독립적인 메모리 공간을 가지기 때문에 안정성이 높지만 프로세스 생성과 컨텍스트 스위칭 비용이 크다. 또한 IPC가 필요해서 파이프, 소켓등을 사용해야 한다. 멀티 스레드는 메모리를 공유할 수 있고 스레드 생성과 컨텍스트 스위칭 비용이 상대적으로 적지만 자원 공유 동기화 문제를 신경써야 하며 이를 위해 락, 세마포어를 많이 걸면 성능이 저하될 수 있다. 웹서버 - 멀티 프로세스. 웹 브라우저(안정성 독립성), 게임 - 멀티스레드(응답성, 효율성)

### 프로세스 동기화에 대해 설명하시오.
다중 프로세스 환경에서 자원에 한 프로세스만이 접근 가능하도록 하는 것. 예를 들어 파일 시스템이나 데이터베이스에 여러 프로세스가 동시에 수정하려고 해서 예상 결과가 다르게 나오는 것을 경쟁 상태라고 하는데 이런 문제를 해결 하는 것이다. 임계 구역을 보호하는 것이 핵심
임계구역 문제 해결을 위한 3가지 조건을 충족해야 한다.
- 상호 배제 (Mutual Exclusion) : 한 프로세스가 임계 구역에 있으면 다른 프로세스 접근 불가
- 진행 (Progress) : 임계 구역에 아무도 없으면 진입 프로세스 결정 후 진입시킴.
- 유한 대기 (Bounded Wating) : Starvation 방지를 위해 다른 프로세스 진입 유한 횟수로 제한


### Thread-safe 하다는 의미와 설계하는 법에 대해 설명하시오.
여러 스레드가 동시에 특정 코드를 실행할 때 해당 코드가 예상된 대로 작동하고 데이터의 일관성이 유지됨을 보장하는 것. 임계 구역을 설정해서 동기화 문제 해결. 이 때 wait 과 notify를 이용해서 무한 대기 하는 것 방지.

### 교착상태와 기아상태의 해결방법에 대해 설명하시오.
교착 상태는 서로 다른 프로세스가 서로 점유하고 있는 자원을 서로 반납을 요구하고 있어서 작업이 멈춘 상태.
다음 4가지 조건을 충족하면 발생한다.
상호 배제 : 한번에 한 프로세스만 해당 자원을 사용할 수 있어야 한다.
점유 대기 : 할당된 자원을 가진 상태에서 다른 자원을 기다린다.
비선점 : 다른 프로세스의 자원을 뺏을 수 없다.
순환대기 : 각 프로세스가 순환적으로 다음 프로세스가 요구하는 자원을 가지고 있다.

-> 해결방법
예방 : 4가지 조건 중 하나라도 만족되지 않도록 한다.
회피 : 알고리즘이 데드락이 발생하지 못하게 한다.
회복 : 교착 상태가 발생하면 해결하는 것을 추가한다.
무시 : 회복과정의 성능 저하가 심하면 무시한다.

기아 상태는 우선순위가 높은 프로세스가 계속해서 자원을 점유해서 다른 프로세스가 무한 대기하고 있는 상태.
-> 우선순위를 수시로 변경시키는 방식으로 해결.

### 세마포어와 뮤텍스의 차이에 대해 설명하시오.
세마포어는 여러 프로세스가 접근 가능한 공유 자원을 관리하는 방식(read-write problem)이고 뮤텍스는 한 자원에 하나의 프로세스만 접근 가능하도록 관리하는 방식(count problem)이다. 세마포어는 뮤텍스가 될 수 있지만 뮤텍스는 세마포어가 될 수 없다.

### 가상 메모리에 대해 설명하시오.
실제 RAM 보다 큰 메모리 공간을 사용하는 것처럼 보이게 하는 기술. 각 프로세스에게 가상 주소 공간을 할당하고 이를 page로 관리한다. 당장 필요한 메모리만 RAM에 올리고 RAM이 꽉차면 page 교체를 이용해서 올리는 방식으로 구현해 사용자 입장에서는 RAM이 무한한 크기처럼 느껴지게 추상화하는 것이다.

### PCB, PC, 캐시메모리
PCB (Process Control Block) : 프로세스에 대한 중요한 정보를 저장한다. 커널 메모리에 잇음
PC (Program Counter) : 프로세스 실행을 위한 다음 명령의 주소를 표시하는 레지스터.
캐시 메모리 : 자주 사용되는 데이터를 CPU에 가까운 메모리에 저장하는 것.(L1, L2, L3)

### 캐시란 무엇이며 캐시의 지역성에 대해 설명하시오.
캐시는 자주 사용하는 데이터를 CPU 칩 내부에 저장하여 작업 처리 속도를 높이는 것이다. 자주 사용하는 데이터에 대한 판별을 위해 지역성 원리가 사용된다. 이는 시간 지역성, 공간 지역성으로 나뉜다. 시간 지역성은 최근에 접근한 데이터에 다시 접근하는 경향성을 말하고 공간 지역성은 최근에 접근한 데이터의 주변에 다시 접근하는 경향성을 말한다.
