### 시간복잡도에 대해 설명하시오.
알고리즘이 문제 해결을 위해 필요한 시간의 척도이다. 빅오표기법으로 표시하며 들어온 데이터 n개에 대한 대략적인 계산양을 n을 이용한 수식으로 나타낸다.

### 배열과 링크드 리스트의 차이에 대해 설명하시오
배열은 메모리상에 순서대로 데이터를 저장하는 자료구조이고 링크드 리스트는 다음 데이터의 위치에 대한 포인터로 연결된 자료구조이다. 배열은 인덱스로 조회할 수 있기 때문에 조회 성능이 높지만 삽입, 삭제의 비용이 많이 든다. 링크드 리스트는 순회해서 조회해야 하지만 삽입 삭제가 용이하다.

### List와 Set의 차이에 대해 설명하시오.
리스트는 중복된 데이터를 저장하고 순서를 유지하는 선형 자료구조이고 셋은 중복되지 않은 데이터를 저장하고 순서를 유지하지 않는 선형 자료구조이다.

### Hash Function, Hash Table에 대해 설명하시오
해쉬 함수는 임의 크기의 데이터를 고정 크기의 값으로 매핑하는 함수이다. 해쉬 테이블은 해쉬 함수를 사용해서 데이터를 저장하는 자료구조로 해쉬값에 해당하는 버킷을 찾아 반환하는 key-value 형식의 자료구조이다. 검색 속도가 매우 빠르다.

### 스택과 큐에 대해 설명하시오.
스택은 나중에 들어온 데이터가 먼저 나가는 LIFO 방식의 자료구조이고 큐는 먼저 들어온 데이터가 먼저 나가는 순차적인 FIFO 방식의 자료구조이다.

### Heap, Priority Queue에 대해 설명하시오.
힙은 완전이진트리 자료구조로 최소 힙과 최대 힙이 존재하며 최소 힙은 자식 노드보다 부모 노드의 값이 작거나 같아야 한다. 추가 삭제 log n 소요되고 최소값을 찾는 것은 log(1)이 소요된다. 우선순위 큐는 우선순위가 높은 순서대로 나가는 큐로 힙은 우선순위 큐를 구현하는데 사용된다.

### Tree, Binary Tree, BST, AVL Tree에 대해 설명하시오.
트리는 노드들이 간선으로 연결된 비선형 자료구조를 말한다. 루트 노드에서 시작해서 부모-자식 간계로 연결되어 있다. 사이클이 없는 연결된 그래프며 계층적 구조를 가진다.
이진트리는 각 노드가 최대 두개의 자식을 가질 수 있는 트리를 말한다.
BST(Binary Search Tree) : 부모의 왼쪽 자식노드는 항상 부모노드보다 작고 아른쪽 자식노드는 항상 부모노드보다 크다. 1부터 10까지 순차적으로 저장하면 이 형태는 리스트와 같아지고 시간복잡도는 O(n)이 된다.
AVL : 위의 문제를 해결하기 위한 균형잡힌 이진 검색 트리로 모든 노드의 왼쪽 서브트리와 오른쪽 서브트리의 높이 차이가 최대 1인 특성을 가진다. 이는 logn의 시간복잡도를 보장한다.

### 피보나치 수열을 코드로 구현할 때
가장 처음으로 생각할 수 있는 방식은 재귀 호출로 구현하는 방식이다. 이 때 매번 연산이 수행되어야 하기 때문에 오버헤드가 소요되는데 중복된 연산을 메모리에 저장해두고 결과가 존재하지 않을 때만 연산을 수행하는 DP 방식으로 구현하면 더 빠르게 동작한다.

### DFS, BFS에 대해 설명하시오.
둘 다 완전 탐색을 위한 알고리즘이다. DFS는 깊이 우선 탐색으로 트리의 노드를 따라 리프 노드까지 탐색한 뒤에 더이상 내려갈 수 없으면 돌아와서 다시 갈 수 있는 방향으로 탐색을 시도하는 방식이다. 프루닝으로 유망하지 않은 가지들을 잘라내어 연산을 줄일 수 있으며 스택이나 재귀를 사용하여 구현할 수 있다. BFS는 넓이 우선 탐색으로 루트 노드로부터 출발하여 인접한 노드들을 모두 탐색후 다음 깊이로 내려가는 방식이다. 최단 거리 등을 찾을 때 사용하면 유리하며 큐로 구현한다.
